```python
dataset = pd.read_csv("/content/Social_Network_Ads.csv")
x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

```python
classifier = GaussianNB()
classifier.fit(x_train_sc, y_train)
```
